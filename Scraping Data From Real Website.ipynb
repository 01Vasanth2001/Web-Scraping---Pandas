```python
#importing libraries
from bs4 import BeautifulSoup
import requests
url = "https://en.wikipedia.org/wiki/List_of_universities_in_India"
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
print(soup)
```

specify what data we're looking for!

we are going to need only a specific data from that website

```python
soup.find_all('table')
```

among the above tables we are going to need only the player's all kind of formats

```python
soup.find_all('table', class_='wikitable')[0]
```

```python
#let
table = soup.find_all('table',class_='wikitable')[0]
```

```python
table.find_all('th')
```

```python
#let
state_titles = table.find_all('th')
```

```python
# to show only the words
[title.text for title in state_titles]
```

```python
#without'\n'
[title.text.strip() for title in state_titles]
```

```python
#and let
state_titles = [title.text.strip() for title in state_titles] 
```

```python
# importing pandas and printing the above data in form of a data frame
import pandas as pd
df = pd.DataFrame(columns = state_titles)
df
```

```python
#printing data under each title
column_data = table.find_all('tr')
```

```python
#printing data under each title
table.find_all('tr')
```

```python
#let
column_data = table.find_all('tr')
```

```python
#print all td within that tr
for row in column_data:
    print(row.find_all('td'))
```

```python
#for precise printing of individual data under each title, tyhe python script is as follows!
for row in column_data:
    row_data  = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```

```python
# we want the above data to be in form of a dataframe, to do that,, follow up in the next script
for row in column_data[1:]:       #[1:] is to eliminate the first empty elements
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```

prints all data in logical orders

```python
#to be in data frame
for row in column_data[1:]:       
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = individual_row_data
   
```
